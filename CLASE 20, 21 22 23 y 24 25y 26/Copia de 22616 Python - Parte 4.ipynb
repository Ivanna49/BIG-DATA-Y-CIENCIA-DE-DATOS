# Listas

```python
mar = ['tiburon', 'calamar', 'delfin', 'pulpo']
print(mar)
```

## Modificar un elemento

```python
print(mar)

mar[1] = 'Orca'
print(mar)
```

## Manejar varios elementos a la vez

```python
print(mar[1], mar[2])
print(mar[1:3]) # Imprimo del elemento [1] al elemento [2] ( 3 - 1)
```

```python
print(mar[1:]) # Imprimo desde el elemento [1] hasta el final
```

```python
print(mar[:2]) # imprimo desde el principio hasta el elemnto 1
```

```python
print(mar)
```

```python
num = [1,2,3,4,5,6,"matias",8,9,10,11,12,13,14,15]
print(num[1:11:2])
```

```python
print(num[::2])
```

## Metodos de Listas

### Lista.append()

```python
pez = ['dorado', 'salmon', 'bacalo', 'pejerey']
print(pez)

pez.append('Mojarrita')
print(pez)
```

```python
pez2 = ['dorado', 'salmon', 'bacalo', 'pejerey']
print(pez2+['mojarrita'])
```

### Lista.insert()

```python
pez.insert(2, 'anchoa')
print(pez)
```

### Lista.extend()



```python
mas_peces = ['Globo', 'Payaso']

pez.extend(mas_peces)
print(pez)
```

```python
pez.insert(3, ['merluza', 'Espada'])
print(pez)
```

```python
pez.remove('Globo')
print(pez)
```

```python
pez.pop(2)
print(pez)
```

```python
pez.index('salmon')
```

```python
pez.count('pejerey')
```
