###while
Se ejecuta el mismo bloque de código reiteradas veces, mientras una condicón permanece verdadera

**IMPORTANTE: ** me tengo que asegurar de crear las condiciones para que la expresión del while cambie su valor de verdad en algún momento para poder salir del bucle. (BUCLE INFINITO)


```python
num = 5

while num < 10:
  print(num)
  num += 1
```

```python
respuesta = ""

while respuesta != "salir":
  print("Gracias por quedarte!")
  respuesta = input("ingresa un comando ('salir' para finalizar): ")

print("chau")
```

###for
Recorre un objeto iterable o repite un fragmento de código una determinada cantidad de veces (método range)

```python
mi_lista = ["Matias", 28, 1234567, "CABA", True]

for e in mi_lista:
  print(e)

# e = mi_lista[]
```

```python
for i in range(3, 10, 2):
  print(i)
```

El método que len() me indica la cantidad de elementos de mi objeto iterable (string, listas, tuplas o diccionarios)

```python
mi_lista = ["Matias", 28, 1234567, "CABA", True, "Soltero"]
longitud = len(mi_lista)

for i in range(0, longitud):
  print(mi_lista[i])
```

```python
mi_lista = ["Matias", 28, 1234567, "CABA", True, "Soltero"]

for i in range(0, len(mi_lista)):
  print(mi_lista[i])
```

###Breakpoints
Rompen con el comportamiento de un bucle o función

###pass
Devuelve NULL al leer el interior de un bucle. Se usa en for, while, condicionales, funciones, clases



```python
for i in range(10):
  if i == 3:
    pass               # DESPUES COMPLETAR
  print(i)

print("fin")
```

###break
Termina el bucle actual y salta a la siguiente instrucción del programa


```python
for i in range(10):
  if i == 3:
    break               
  print(i)
  
print("fin")
```

###continue
Salta a la siguiente iteración del bucle, ignorando la actual


```python
for i in range(10):
  if i == 3:
    continue               
  print(i)
  
print("fin")
```

# Paramentros Indefinidos



```python
def sumar(*args):   #args es una tupla
  suma = 0
  for i in args:
    suma += i
  print(suma)  

sumar()
sumar(1,5,9)
sumar(1,6,80,125,654,1586)
```

```python
def imprimir(*pepe):
  for i in pepe:
    print("hola", i)

imprimir("matias")
imprimir("Matias", "Marcos", "Lucas")
```
